### Problem Statement  : Predict percentage of students based on the number of study hours. What will be the predicted score if a student studies for 9.25 hrs/day








### Import Libraries


```python
# Importing all libraries required in this notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Loading Dataset


```python
# Reading data from remote link
url='http://bit.ly/w-data'
data=pd.read_csv(url)
print("Data imported successfully")
```

### Exploratory Data Analysis


```python
#first five rows of dataframe
data.head()
```

```python
#concise summary of dataframe
data.info()
```

```python
#description of data
data.describe()
```

```python
#Checking correlation
data.corr(method='pearson')
```

```python
#checking for null values
data.isnull().sum()
```

### Visualizing Data

```python
# Plotting the distribution of scores
data.plot(x='Hours', y='Scores', style='o', c='r')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show
```

### Preparing the data

```python
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values 
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0) 
```

### Training the model

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 

print("Training complete.")
```

### Plotting  the Regression Line

```python
# Plotting the regression line
line = regressor.coef_*X+regressor.intercept_
#plot for the test data
plt.scatter(X, y)
plt.plot(X, line, label='Line of best fit', color='g');
plt.show()
```

### Making Predictions

```python
print(X_test) 
y_pred = regressor.predict(X_test) 
```

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

```python
# Plotting the Bar graph to depict the difference between the actual and predicted value

df.plot(kind='bar',figsize=(11,7) ,colormap='Paired')
plt.grid(which='major', linewidth='1', color='green')
plt.show()
```


### Evaluating Models

```python
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred)) 
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
```

### Test with your own data

```python
hours = 9.25
own_pred = regressor.predict([[hours]])
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```
